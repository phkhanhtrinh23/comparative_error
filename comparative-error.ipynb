````python
import re
from nltk.corpus import wordnet as wn
from nltk.stem.wordnet import WordNetLemmatizer
from pattern.en import comparative

data, labels = [], []

with open("data.txt","r") as file1:
  while True:
    line = file1.readline()
    if not line:
        break
    data.append(line.strip("\n"))
file1.close()

with open("labels.txt","r") as file2:
  while True:
    line = file2.readline()
    if not line:
        break
    labels.append(line.strip("\n"))
file2.close()

# Regular expression
def preprocess(text):
  # text = re.sub(r'[^\w\s]', '', text) # remove everything except words and space
  text = text.split(' ')
  return text

new_data = []
wnl = WordNetLemmatizer()
for sentence,label in zip(data,labels):
  print(f'O: {sentence}')
  tokens = preprocess(sentence)
  new_string = ""
  for i in range(len(tokens)-1,-1,-1):
    flag = False
    if tokens[i] == 'than':
      pos = i
      while pos >= 0:
        real_token = wnl.lemmatize(tokens[pos], 'a')
        for s in wn.synsets(real_token):
          if s.name().split('.')[0] == real_token and s.pos() in ['a','s']: # Check if this word is and Adjective
            flag = True
            tmp = comparative(real_token) # Convert to comparative form
            if tokens[pos-1] == 'more':
              if tokens[pos-1] + ' ' + tokens[pos] != tmp:
                tokens[pos-1] = ''
                tokens[pos] = tmp
            elif tokens[pos] != tmp:
                tokens[pos] = tmp
            break
        if flag == True:
          break
        pos -= 1
    if i == len(tokens)-1:
      new_string = tokens[i]
    else:
      new_string = tokens[i] + " " + new_string if tokens[i] != '' else new_string
  print(f'L: {label}')
  print(f'P: {new_string}')
  print()
  new_data.append(new_string)
````
